# K-Folds

```python
def split(k):
    temp1=df[df['Label']==1.0]
    new1=np.array_split(temp1,k)
    
    temp2=df[df['Label']==-1.0]
    new2=np.array_split(temp2,k)
    
    return new1,new2
```

```python
k=4
new1,new2=split(k)
```

```python
def k_folds():
    all_folds= new1
    for i in range(0,4):
         all_folds[i]=all_folds[i].append(new2[i])
    return all_folds
```

```python
all_folds=k_folds()
```

```python
train_folds=[[0,1,2],[1,2,3],[0,3,1],[0,2,3]]
test=[3,0,2,1]
```

```python
def K_Folds_Processing(k):
    x_train=pd.DataFrame()
    y_train=[]
    x_test=[]
    y_test=[]
    for i in range(0,k):
    #     Making Training and testing Folds
        print("\t\tK-Fold #",i+1)
        index=train_folds[i]
        j=index[0]
        k=index[1]
        l=index[2]
        x_train=pd.concat([all_folds[j],all_folds[k],all_folds[l]])
        y_train=x_train['Label']
        x_train.drop(labels='Label',axis=1,inplace=True)

        x_test=all_folds[test[i]]
        y_test=x_test['Label']
        x_test=x_test.loc[:, x_test.columns != 'Label']

    #     Model
        clf = tree.DecisionTreeClassifier()
        clf = clf.fit(x_train, y_train)

        y_pred=clf.predict(x_test)
        print('Accuracy:', accuracy_score(y_test, y_pred)*100)
        print("F1 Score : ",f1_score(y_test, y_pred, average='macro')*100)
        print('\n')

    
```

```python
K_Folds_Processing(k)
```

# SK Learn 

```python
labels=df['Label']
df.drop(labels='Label',axis=1,inplace=True)
```

```python
kfolds = KFold(n_splits=4)
kfolds.get_n_splits(df)
i=0
for train_index, test_index in kfolds.split(df):
    print("\t\t K-Fold#",i+1)
    X_train, X_test = df.iloc[train_index], df.iloc[test_index]
    y_train, y_test = labels.iloc[train_index], labels.iloc[test_index]
    
    clf = tree.DecisionTreeClassifier()
    clf = clf.fit(X_train,y_train)
    
    y_pred=clf.predict(X_test)
    print('Accuracy:', accuracy_score(y_test, y_pred)*100)
    print("F1 Score : ",f1_score(y_test, y_pred, average='macro')*100)
    i+=1
```
